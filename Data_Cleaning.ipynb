## Rectifying Zip Code Column

```python
#Listing the values in the zip codes column
zip_codes = losdata_v2['Zip Code - 3 digits'].unique().tolist()
print(zip_codes)
```

```python
#Including normalization to unnecessary duplicated float values
def decimal_elimination(value):
  try:
      return str(float(value)).rstrip('.0')
  except ValueError:
      return value

#Applying the normalization function
losdata_v2['Zip Code - 3 digits'] = losdata_v2['Zip Code - 3 digits'].apply(decimal_elimination)

#Listing the unique values in the zip codes columns again
zip_codes = losdata_v2['Zip Code - 3 digits'].unique().tolist()
print(zip_codes)
```

```python
#Re-editing the values
def digit_correction(zip):
  try:
      zip = int(zip)
      if 0 < zip < 9:
        return str(zip*100)
      elif 10 < zip < 15:
        return str(zip*10)
      else:
        return str(zip)
  except ValueError:
        return str(zip)

#Applying the normalization function
losdata_v2['Zip Code - 3 digits'] = losdata_v2['Zip Code - 3 digits'].apply(digit_correction)

#Listing the unique values in the zip codes columns again
zip_codes = losdata_v2['Zip Code - 3 digits'].unique().tolist()
print(zip_codes)
print(losdata_v2['Zip Code - 3 digits'].dtype)
null_zip = losdata_v2['Zip Code - 3 digits'].isnull().sum()
print(f"The null values left in the zip codes column: {null_zip}")
```

```python
#Checking the values of the zip codes column
zip_codes = losdata_v2['Zip Code - 3 digits'].unique().tolist()
print(zip_codes)
```

```python
#Checking the number of rows with 'nan' zip code values
nan_count = losdata_v2['Zip Code - 3 digits'].value_counts().get('nan', 0)
print(f"Number of rows with nan values in the Zip Code column: {nan_count}")
```

```python
#Eliminating 'nan' values
losdata_v3 = losdata_v2[losdata_v2['Zip Code - 3 digits'] != 'nan']
display(losdata_v3.head(10))
```

# Target Variable Conversion

```python
#The conversion of LoS to a numerical variable is done as a classification task with around 120 categories can seem not ideal, hence a conversion is to be made and the task is made to be one of regression.

#Converting Length of Stay to a numerical column
losdata_v5['Length of Stay'] = losdata_v5['Length of Stay'].str.replace('120 +', '120')
losdata_v5['Length of Stay'] = pd.to_numeric(losdata_v5['Length of Stay'])
display(losdata_v5['Length of Stay'].head(10))
```

# Outlier/Anomaly Detection

```python
#Outlier/Anomaly Detection for Total Charges is done due to abnormally extreme values in both ends of the data

#IQR Method for Total Charges
tcq1 = losdata_v5['Total Charges'].quantile(0.25)
tcq3 = losdata_v5['Total Charges'].quantile(0.75)
iqr_tc = tcq3 - tcq1
tc_outliers = losdata_v5[(losdata_v5['Total Charges'] < (tcq1 - 1.5 * iqr_tc)) | (losdata_v5['Total Charges'] > (tcq3 + 1.5 * iqr_tc))]
tc_out_dim = tc_outliers.shape
print(f"Total Charges outliers based on IQR: {tc_out_dim[0]}")
print(f"The percentage of rows with Total Charges outliers: {tc_out_dim[0]/(losdata_v5.shape[0])*100}%")
```

## Anomaly Rectification

```python
#Importing IsolationForest
from sklearn.ensemble import IsolationForest

#Applying IsolationForest for Total Charges
iso_f = IsolationForest(contamination=0.085, random_state=34) #Contamination defined to be 0.085 due to the estimated percentage of outliers for Total Charges being 8.5%
losdata_v5['Anomaly Score'] = iso_f.fit_predict(losdata_v5[['Total Charges']])
print('The value counts for anomalies in the column is given to be')
print(losdata_v5['Anomaly Score'].value_counts())
tch_anomaly = losdata_v5[['Total Charges', 'Anomaly Score']]
display(tch_anomaly.sort_values(by='Total Charges', ascending=True).head(10))
```

```python
#Clearing out the anomalies
losdata_v6 = losdata_v5[losdata_v5['Anomaly Score'] == 1]
losdata_v6 = losdata_v6.drop(columns=['Anomaly Score'])
display(losdata_v6.describe())
display(losdata_v6.shape)
```

# Eliminating Duplicates in Operating Certificate Number

```python
#Identifying whether there are duplicates in 'Operating Certificate Number'
unique_cert = losdata_v6.groupby('Facility Name')['Operating Certificate Number'].nunique()
duplicates = unique_cert[unique_cert > 1]
dup_df = duplicates.to_frame()
dup_df['Values'] = losdata_v6.groupby('Facility Name')['Operating Certificate Number'].unique()
display(dup_df)
```

```python
#Adding more columns to try single out the values to make grounds for imputation
op_count = losdata_v6['Operating Certificate Number'].value_counts()
dup_df['Occurences'] = dup_df['Values'].apply(lambda x: [op_count[num] for num in x])
dup_df['County'] = losdata_v6.groupby('Facility Name')['Hospital County'].unique()
dup_df['Area'] = losdata_v6.groupby('Facility Name')['Health Service Area'].unique()
display(dup_df)
```

# Checking between Client Classification Descriptions and Codes

```python
#This is done to check whether client classification codes truly and uniquely represent each every decription tied to it to analyse for potential redundance
code_desc_pairs = [('CCS Diagnosis Code', 'CCS Diagnosis Description'), ('CCS Procedure Code', 'CCS Procedure Description'), ('APR DRG Code', 'APR DRG Description'), ('APR MDC Code', 'APR MDC Description'), ('APR Severity of Illness Code', 'APR Severity of Illness Description')]
for code_col, desc_col in code_desc_pairs:
  unique_desc_per_code = losdata_v6.groupby(code_col)[desc_col].nunique()
  non_unique = unique_desc_per_code[unique_desc_per_code > 1]
  if not non_unique.empty:
    print(f"Inconsistent pairings found between {code_col} and {desc_col}\n")
    print(non_unique)
  else:
    print(f"\nNo inconsistent pairings found between {code_col} and {desc_col}")
```

# Feature Elimination

```python
#Taking out a few columns that doesn't seem to provide much to the analysis
losdata_v7 = losdata_v6.drop(columns=['Discharge Year', 'Birth Weight', 'Abortion Edit Indicator', 'Facility ID', 'index'])
display(losdata_v7.head(15))
print(losdata_v7.shape)
```

# T-Tests comparing Elective and Emergency Admissions based on average LoS

```python
#Executing T-Test to compare the means of Length of Stay pertaining to both elective and emergency admissions
from scipy.stats import ttest_ind

elective_vals = losdata_v7[losdata_v7['Type of Admission'] == 'Elective']['Length of Stay']
emergency_vals = losdata_v7[losdata_v7['Type of Admission'] == 'Emergency']['Length of Stay']
t_stat, p = ttest_ind(elective_vals, emergency_vals, equal_var=False)
print(f"T-Statistic for the test: {t_stat}")
print(f"P-Value: {p}")
if p < 0.05:
  print('The difference between the categories are statistically significant')
else:
  print('The difference between the categories are not statistically significant')
#The results of this test points out the significance in analysing both admission types against LoS
```
