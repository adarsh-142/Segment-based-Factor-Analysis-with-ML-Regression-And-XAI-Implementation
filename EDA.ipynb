# Variable Correlations to LoS

## Categorical Variable Correlations to LoS

```python
#To identify the extent of correlation of categorical variables against LoS using correlation ratio(eta squared)

#Importing numpy for arithmetic use
import numpy as np

#Setting up dataframes for correlation purposes
elective_corr1 = elective.drop(columns=['APR Severity of Illness Description'])
emergency_corr1 = emergency.drop(columns=['APR Severity of Illness Description'])

#Removal employed as the numerical code representing the severity of illness column exist and will be used for the numerical correlation analysis

#Implementing correlation ratio using a function
def correlation_ratio(df, df_name):
  ratios = []
  for col in df.columns:
    if df[col].dtype == 'object':
      mean_los = df['Length of Stay'].mean()
      group_means = df['Length of Stay'].groupby(df[col]).mean()
      group_sizes = df[col].value_counts()
      bgv = ((group_means - mean_los) ** 2 * group_sizes).sum()
      wgv = df['Length of Stay'].var(ddof=0)*len(df['Length of Stay'])
      corr_ratio = bgv/ (bgv + wgv)
      ratios.append({f'{df_name} Columns': col, 'Correlation Values': corr_ratio})
  ratios_df = pd.DataFrame(ratios)
  return ratios_df.sort_values(by='Correlation Values', ascending=False)

#Implementing function calls onto subset dataframes
elective_ratios = correlation_ratio(elective_corr1, 'Elective')
emergency_ratios = correlation_ratio(emergency_corr1, 'Emergency')
display(elective_ratios)
display(emergency_ratios)
```

## Numerical Variable Correlations to LoS

```python
#Implementing a correlation matrix based on the Pearson's Correlation Coefficient as default

#Importing a library to implement custom colors for the correlation matrix
from matplotlib.colors import LinearSegmentedColormap

#Correlation Matrix
elective_corr2 = elective.select_dtypes(include=['int64', 'float64'])
emergency_corr2 = emergency.select_dtypes(include=['int64', 'float64'])

#Creating a custom colormap for correlation indication
colors = ['red', 'yellow', 'green']
n_bins = 100
cmap_name = 'red_yellow_green'
custom_cmap = LinearSegmentedColormap.from_list(cmap_name, colors, N=n_bins)

#Creating a function for correlation matrix for both datasets
def correlation_matrix(df, adm_type):
  cor_matrix = df.corr()
  plt.figure(figsize=(8,6))
  sns.heatmap(cor_matrix, annot=True, cmap=custom_cmap, vmin= -1, vmax= 1)
  plt.title(f'Correlation Matrix of Numerical Variables in {adm_type} Admissions')
  corr_los = cor_matrix['Length of Stay'].drop('Length of Stay').sort_values(ascending=False)
  return corr_los

#Function call for each dataframe
elec_corr_df = correlation_matrix(elective_corr2, 'Elective')
display(elec_corr_df)
emer_corr_df = correlation_matrix(emergency_corr2, 'Emergency')
display(emer_corr_df)
```
